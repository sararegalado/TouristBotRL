```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification", 
                       model="Recognai/bert-base-spanish-wwm-cased-xnli")
```

## Definir las categorías de intención (fuentes de datos)

Para el chatbot turístico, vamos a clasificar las consultas del usuario en 3 categorías según la fuente de datos apropiada:

```python
# Definir categorias
candidate_labels = {"Museo":1, "Restaurante":2, "Tienda":3, "Ayuntamiento":4}
```

## Preguntas de competencia

```python
ordenes = ["quiero ir al museo pero no al restaurante", "busca un restaurante", "por favor ayudame a llegar al ayuntamiento", "buscame tiendas que esten cerca de restaurantes"]
```

## Clasificación Zero-Shot de las consultas

```python
# Clasificar cada consulta - USANDO EL CLASIFICADOR CORRECTO
print("=" * 100)
print("CLASIFICACIÓN DE INTENCIONES PARA ROUTING A FUENTES DE DATOS")
print("=" * 100)

predictions = []
labels = list(candidate_labels.keys())

for query in ordenes:
    # IMPORTANTE: usar 'classifier' en lugar de 'pipe'
    result = classifier(query, labels)
    
    # El resultado contiene las etiquetas ordenadas por score
    top_label = result['labels'][0]
    top_score = result['scores'][0]

    
    print(f"\nConsulta: {query}")
    print(f"Intención: {top_label}")
    print(f"Confianza: {top_score:.2%}")
    print(f"Accion: {candidate_labels.get(top_label)}")

    predictions.append(top_label)
    
    # Mostrar warning si la confianza es baja
    if top_score < 0.60:
        print(f"Confianza baja - Top 3: {dict(zip(result['labels'][:3], [f'{s:.2%}' for s in result['scores'][:3]]))}")
    print("-" * 100)
```
