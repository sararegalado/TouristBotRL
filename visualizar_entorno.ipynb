```python
# Función para visualizar el grid con coordenadas correctas
def visualizar_grid(grid):
    """
    Visualiza el grid 4x4 donde (1,1) está en la esquina inferior izquierda
    """
    fig, ax = plt.subplots(figsize=(8, 8))
    
    # Colores: 0=negro (vacío), 1=amarillo (tienda), 2=naranja (restaurante)
    colores = {
        0: 'black',     # vacío
        1: 'yellow',    # tienda
        2: 'orange'     # restaurante
    }
    
    nombres = {
        0: '',
        1: 'TIENDA',
        2: 'REST'
    }
    
    # Dibujar cada celda
    # Invertimos las filas para que fila 0 del array esté arriba
    for i in range(4):  # filas del array (0-3)
        for j in range(4):  # columnas del array (0-3)
            valor = grid[i, j]
            color = colores[valor]
            
            # Coordenadas de usuario: (col+1, 4-row)
            # col+1: convierte índice 0-3 a coordenadas 1-4
            # 4-i: invierte las filas para que (1,1) esté abajo-izquierda
            coord_usuario_col = j + 1
            coord_usuario_fila = 4 - i
            
            # Dibujar rectángulo
            rect = Rectangle((j, i), 1, 1, 
                           facecolor=color, 
                           edgecolor='white', 
                           linewidth=3)
            ax.add_patch(rect)
            
            # Añadir nombre del objeto si no está vacío
            if valor != 0:
                ax.text(j + 0.5, i + 0.5, nombres[valor],
                       ha='center', va='center', 
                       fontsize=12, color='black', weight='bold')
            
            # Añadir coordenadas de usuario en cada celda
            ax.text(j + 0.5, i + 0.8, f'({coord_usuario_col},{coord_usuario_fila})',
                   ha='center', va='top', 
                   fontsize=8, color='white', weight='bold',
                   bbox=dict(boxstyle='round', facecolor='black', alpha=0.7))
    
    # Configurar ejes
    ax.set_xlim(0, 4)
    ax.set_ylim(0, 4)
    ax.set_aspect('equal')
    
    # Invertir el eje Y para que (1,1) esté abajo
    ax.invert_yaxis()
    
    # Etiquetas de los ejes con coordenadas de usuario
    ax.set_xticks([0.5, 1.5, 2.5, 3.5])
    ax.set_xticklabels(['1', '2', '3', '4'], fontsize=12, weight='bold')
    ax.set_yticks([0.5, 1.5, 2.5, 3.5])
    ax.set_yticklabels(['4', '3', '2', '1'], fontsize=12, weight='bold')
    
    ax.set_xlabel('Columna', fontsize=14, weight='bold')
    ax.set_ylabel('Fila', fontsize=14, weight='bold')
    ax.set_title('Grid 4x4 - (1,1) en esquina inferior izquierda', 
                 fontsize=16, weight='bold', pad=20)
    
    # Grid
    ax.set_xticks(range(5), minor=True)
    ax.set_yticks(range(5), minor=True)
    ax.grid(which='minor', color='white', linewidth=2)
    
    plt.tight_layout()
    plt.show()
    
    # Información de los objetos
    print("\n" + "="*60)
    print("OBJETOS EN EL GRID:")
    print("="*60)
    for i in range(4):
        for j in range(4):
            if grid[i, j] != 0:
                coord_col = j + 1
                coord_fila = 4 - i
                nombre = nombres[grid[i, j]]
                print(f"{nombre:10} en posición (columna={coord_col}, fila={coord_fila})")

# Visualizar el grid
visualizar_grid(grid)
```
